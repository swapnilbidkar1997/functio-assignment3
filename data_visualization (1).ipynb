3. create a dataset representing categories and their corresponding  values . compare different categories based on numerical  values.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {'Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
        'Value': [10, 15, 12, 18, 20, 8, 11, 14, 16]}
df = pd.DataFrame(data)
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Comparison of Categories based on Value')
plt.show()

sns.violinplot(x='Category', y='Value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Comparison of Categories based on Value (Violin Plot)')
plt.show()
```

4. generate a dataset  with categories and numerical values. visualize the distribution of a numerical variables across different categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
data = {'Category': ['A'] * 50 + ['B'] * 50 + ['C'] * 50,
        'Value': np.concatenate([np.random.normal(10, 2, 50),
                                 np.random.normal(15, 3, 50),
                                 np.random.normal(20, 4, 50)])}
df = pd.DataFrame(data)
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Distribution of Value across Categories (Box Plot)')
plt.show()
sns.violinplot(x='Category', y='Value', data=df)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Distribution of Value across Categories (Violin Plot)')
plt.show()
```

5.  generate a synthetic dataset with correlated features. visualize the correletion matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
num_samples = 100
feature1 = np.random.rand(num_samples)
feature2 = 2 * feature1 + np.random.rand(num_samples)
feature3 = 0.5 * feature1 + np.random.rand(num_samples)
data = {'Feature1': feature1, 'Feature2': feature2, 'Feature3': feature3}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

4 using the given x and y data ,generate a 3d surface ploat to visuallize the function z = sin * (underroot x^2+y^2)


```python


import plotly.graph_objects as go
import numpy as np
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x, y)


Z = np.sin(np.sqrt(X**2 + Y**2))


fig = go.Figure(data=[go.Surface(x=x, y=y, z=Z)])
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

2.creat a bokeh scatter plot using randomly generated x and y values. use different sizes and colors for the markers based on the sizes and colors colomns


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
output_notebook()
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(10, 20, n)
colors = np.random.choice(['red', 'green', 'blue'], n)
df = pd.DataFrame({'x': x, 'y': y, 'size': sizes, 'color': colors})
p = figure(title="Scatter Plot with Sizes and Colors", x_axis_label='x', y_axis_label='y')
p.scatter(x='x', y='y', source=df, size='size', color='color', legend_field='color')
show(p)

```
